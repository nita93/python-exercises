```python
def check_columns(matrix):
    for x in range(3):
        if matrix[0][x] == matrix[1][x] and matrix[1][x] == matrix[2][x] and matrix[0][x] != 0:
            return matrix[0][x]
    return 0;

def check_rows(matrix):
    for x in range(3):
        if matrix[x][0] == matrix[x][1] and matrix[x][1] == matrix[x][2] and matrix[x][0] != 0:
            return matrix[x][0]
    return 0;
    
def check_diagonal(matrix):
    if ((matrix[0][0] == matrix[1][1] and matrix[1][1] == matrix[2][2]) or (matrix[0][2] == matrix[1][1] and matrix[1][1] == matrix[2][0])) and matrix[1][1] != 0:
        return matrix[1][1]
    else:
        return 0;
    
# function to check the result of the game
# 0 => no winner

def check_tic_tac_toe(matrix):
    column = check_columns(matrix)
    if column != 0:
        return column
    row = check_rows(matrix)
    if row != 0:
        return row
    diagonal = check_diagonal(matrix)
    if diagonal != 0:
        return diagonal
    else:
        return 0
```

```python
# prepare emtpy board

board = [[0,0,0],
        [0,0,0],
        [0,0,0]]

is_player1 = True
player = ""
character = ""
x, y = (1,1)

while check_tic_tac_toe(board) == 0:
    if is_player1:
        player = "Player 1"
        character = "X"
    else:
        player = "Player 2"
        character = "O"
    while True:
        user_choice = input(f"{player}, give me coordinates separated with comma (raw,column): ").split(",")
        x, y = int(user_choice[0]), int(user_choice[1])
        if x >= 1 and x <=3 and y <= 3 and y >= 1:
            if board[x-1][y-1] != 0:
                print("This position is already chosen. Choose another position!")
            else:
                board[x-1][y-1] = character
                is_player1 = not is_player1
                break
        else:
            print("Wrong coordinates. They shouldn't be greater than 3 and lower than 1. Try again!")
    print(board[0])
    print(board[1])
    print(board[2])

if check_tic_tac_toe(board) == 'X':
    print("Player 1 won!")
else:
    print("Player 2 won!")
```
